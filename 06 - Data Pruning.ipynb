# Your Model is Only as Good as the Data it Learns From

## Your Data Contains Bugs/Outliers

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame([[1,2],[3,6],[5,10],[6,12],[7,14],[8,1],[8,16]])
# with column names for ease of use
df.columns = ['Unique User Agents','Unique IPs']
df.sample(1)
```

```python
df.plot()
```

The college campus has thousands of users, but only 1 outbound IP addresses. This is an example of an outlier that will fool your model into thinking that this is a normal event that has the same wieght as everything else. Before running ML on your data, find the outliers and decide at what point you want to cut off and restrict the anomalies to create a "normalized" set of data.


```python
df = pd.DataFrame([[1,2],[3,6],[5,10],[6,12],[7,14],[8,16]])
df.columns = ['Unique User Agents','Unique IPs']
df.sample(1)
df.plot()
```
